```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
#Reading the data

data_train_events = pd.read_csv('train_events.csv')
data_train_series = pd.read_parquet('train_series.parquet')
```

```python
first_series = data_train_events['series_id'].unique().tolist()[0]

#let's work on 1 series 
data_train_events_sample = data_train_events.loc[data_train_events['series_id'] == first_series]
data_train_series_sample = data_train_series.loc[data_train_series['series_id'] == first_series]
```

```python
#visualize Angles and Enmo in train series
plt.figure(figsize=(20,10))
plt.plot(data_train_series_sample['anglez'])
plt.xlabel('steps')
plt.ylabel('angles')
plt.title('Angles')

plt.figure(figsize=(20,10))
plt.plot(data_train_series_sample['enmo'])
plt.xlabel('steps')
plt.ylabel('enmo')
plt.title('Enmo')
```
